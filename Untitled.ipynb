```python
my_list=[]
path="C:/Users/NINANMM.000/OneDrive - University of Cincinnati/Desktop/stuff/yalefaces/*.*"
for file in glob.glob(path):
    print(file)
    a=mpimg.imread(file,cv2.IMREAD_GRAYSCALE)
    my_list.append(a)
```

```python
label=[]
for i in range(1,16):
    for j in range(1,12):
        label.append(j)
        
print(label)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(), 
                                    tf.keras.layers.Dense(128, activation=tf.nn.relu), 
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
model.compile(optimizer = tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(my_list, label, epochs=5)
```
