```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import matplotlib.image as mpimg
import tensorflow as tf
import numpy as np
```

```python


training_data=[]
training_labels=[]
DATADIR = "C:/Users/NINANMM.000/OneDrive - University of Cincinnati/Desktop/next_test"

CATEGORIES = ["subject01", "subject02","subject03", "subject04", "subject05", "subject06","subject07","subject08","subject09", "subject10",
             "subject11","subject12", "subject13", "subject14", "subject15" ]
IMG_SIZEX=280
IMG_SIZEY=280


for category in CATEGORIES:
    
    path = os.path.join(DATADIR,category)  
    class_num= CATEGORIES.index(category)
    j=0
    for img in os.listdir(path): 
        
        img_array = mpimg.imread(os.path.join(path,img) ,cv2.IMREAD_GRAYSCALE) 
        new_array=cv2.resize(img_array,(IMG_SIZEX,IMG_SIZEY))

        training_data.append(new_array)
        training_labels.append(j)
        j+=1
        
        
training_data=np.array(training_data)
training_labels=np.array(training_labels)
```

```python
print(len(training_labels))
print(training_data.shape)
```

```python
plt.imshow(training_data[1])
print(training_labels[1])
print(training_data[1])
```

```python
training_data = training_data / 255.0
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(280, 280)),
    tf.keras.layers.Dense(1000, activation='relu'),
    tf.keras.layers.Dense(500, activation='relu'),
    tf.keras.layers.Dense(300, activation='relu'),
    tf.keras.layers.Dense(100, activation='softmax')
])
```

```python
type(training_data)
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(training_data, training_labels, epochs=1000)
```
